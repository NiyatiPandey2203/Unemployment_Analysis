```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the unemployment rate data
unemployment_data = pd.read_csv("/content/Unemployment in India.csv")
```

```python
print(unemployment_data.columns)
```

```python
# Convert the date column to datetime format
unemployment_data[' Date'] = pd.to_datetime(unemployment_data[' Date'])
```

```python
# Plot the unemployment rate over time
plt.figure(figsize=(10, 6))
plt.plot(unemployment_data[' Date'], unemployment_data[' Estimated Unemployment Rate (%)'])
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average unemployment rate
average_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].mean()
print("Average Unemployment Rate:", average_unemployment_rate)
```

```python
# Calculate the maximum unemployment rate
max_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].max()
print("Maximum Unemployment Rate:", max_unemployment_rate)
```

```python
# Calculate the minimum unemployment rate
min_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].min()
print("Minimum Unemployment Rate:", min_unemployment_rate)
```

```python
# Calculate the standard deviation of the unemployment rate
std_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].std()
print("Standard Deviation of Unemployment Rate:", std_unemployment_rate)
```
